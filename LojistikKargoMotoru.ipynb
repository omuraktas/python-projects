```python
# Akıllı Lojistik ve Paket Fiyatlandırma Motoru

#Alınacak Kargo Bilgileri
gonderiAdi = input("Lütfen göndermek istediğniz ürünü giriniz:").strip().lower().capitalize()
gonderiKilo = float(input("Lütfen ürünün kilosunu giriniz(Sadece rakam):"))
gonderiBoy = float(input("Lütfen ürünün boyunu giriniz:"))
gonderiEn = float(input("Lütfen ürünün enini giriniz:"))
gonderiYukseklik = float(input("Lütfen ürünün yüksekliğini giriniz:"))
gonderiMesafe = float(input("Lütfen göndermek istediğniz mesafeyi giriniz(Mesafe/km):"))
gonderiTuru = input("(Standart/Ekspres/Vip) \n Standart : Mesafe <= 100 km \n Ekspres +50 Tl : Mesafe <= 500km\nVip +100 : Mesafe > 500 km \nLütfen gönderim türünü giriniz : ").strip().lower().capitalize()
gonderiSadik = input("Sadakat kartınız varmı?(Evet/Hayır)").strip().lower().capitalize()

#Mesafe Tür Kontrolü
gonderiKontrol = {
    "Standart": {"min": 1,"max": 100},
    "Ekspres": {"min": 100,"max": 500}, 
    "Vip": {"min": 500 ,"max": float("inf")}}
   
 #Mesafe Ve Tür Doğruluğu Kontrolü 
if gonderiTuru in gonderiKontrol:
    kontrol = gonderiKontrol[gonderiTuru]
    gonderiMesafeTuru = "True"
    gonderiMesafeTur = True
   
    if gonderiMesafe < kontrol["min"]:
        gonderiMesafeTuru = "False"
        gonderiMesafeTur = False
    elif gonderiMesafe > kontrol["max"]:
        gonderiMesafeTuru = "False"
        gonderiMesafeTur = False
else:
    gonderiMesafeTuru = "False"
    gonderiMesafeTur = False
    
#Mesafe Kontrolü
if gonderiMesafe > 0 and gonderiMesafeTur and gonderiKilo > 0 and gonderiBoy > 0 and gonderiEn > 0 and gonderiYukseklik > 0 :
    #Hacim ve Bölge Bİlgileri
    gonderiHacim = gonderiBoy*gonderiEn*gonderiYukseklik
    if gonderiHacim >= 10000:
        gonderiHacimKatagori = ["Büyük",1.7]
    elif gonderiHacim <= 10000 and gonderiHacim >= 5000:
        gonderiHacimKatagori = ["Orta",1.3]
    else :
        gonderiHacimKatagori = ["Küçük",1]
    if gonderiMesafe <= 100:
        gonderiBolge = {"Carpan" : 1.0,"TuruUcreti" : 0}
    elif gonderiMesafe <= 500:
        gonderiBolge = {"Carpan" : 1.3,"TuruUcreti" : 50} 
    else :
        gonderiBolge = {"Carpan" : 1.7,"TuruUcreti" : 100}
        
    #Ücret Hesaplama
    
    gonderiMin = gonderiKilo*5
    gonderiMaliyet = (gonderiMin*gonderiBolge["Carpan"]*gonderiHacimKatagori[1]) + gonderiBolge["TuruUcreti"]
    if gonderiSadik == "Evet" :
        gonderiMaliyet =   gonderiMaliyet*0.85
    
    #Tüm Bilgileri Sözlüge Aktarma
    
    gonderiBilgileri = {
                       "gonderiKilo" : gonderiKilo,
                       "gonderiHacim" : gonderiHacimKatagori[0],
                       "gonderiMesafe" : gonderiMesafe,
                       "gonderiTuru" : gonderiTuru,
                       "gonderiSadik" : gonderiSadik,
                       "gonderiMaliyet" :  gonderiMaliyet,
                       }
    gonderiBilgileri["gonderiAdi"] = gonderiAdi
    
    #Gönderi Müşteri Ücret Onaylama

    gonderiMaliyetOnay = input(f"Maliyet {gonderiBilgileri['gonderiMaliyet']}\nBu ücreti kabul ediyor musunuz? (Evet / Hayır): ").strip().lower().capitalize() 

    if gonderiMaliyetOnay == "Evet" :
        
        print("=========== KARGO LOJİSTİK FATURASI ===========")
        print(f"Paketin İçeriği: {gonderiBilgileri['gonderiAdi']}")
        print(f"Paketin Türü: {gonderiBilgileri['gonderiTuru']} ")
        print(f"Paketin Hacimi: {gonderiBilgileri['gonderiHacim']}")
        print(f"Paketin Kilosu: {gonderiBilgileri['gonderiKilo']} Kg")
        print(f"Paketin Mesafesi: {gonderiBilgileri['gonderiMesafe']} Km")
        if gonderiSadik == "Evet" :
            print(f"Sadık Müşteri : {gonderiBilgileri['gonderiSadik']} (%15 İndirim Uygulandı)")
        print(f"Paketin Maliyeti: {gonderiBilgileri['gonderiMaliyet']} TL")
        print("===============================================")
    else :
        print("İşlem kullanıcı tarafından iptal edildi.\nFatura oluşturulmadı.")
        
        
else :
    if gonderiMesafeTuru == "False" :
       print("Lütfen türü ile mesafeyi doğru!!!") 
    else:
        print("Lütfen negatif veya 0 girmeyin!!!")
```
